```python
%cd /content/drive/MyDrive/
!ls
data=pd.read_csv("transactionsForML.csv",index_col=False)
# test_data=pd.read_csv("test_transactions.csv",index_col=False)
labels=pd.read_csv("train_labels.csv",index_col=False)
# test_labels=pd.read_csv("test_labels.csv",index_col=False)
# test_data.shape
data.head(-11)

data.columns
```

```python
data=data.drop('code',axis=1)
data=data.drop('parent_code',axis=1)
data=data.drop('name',axis=1)
data.head()
```

```python
labels
```

```python

labels.rename(columns={'id':'Id'}, inplace=True)
# test_labels.rename(columns={'id':'Id'}, inplace=True)
merged_df = pd.merge(data, labels, on=['Id'], how='left')
# test_merged_df = pd.merge(test_data, test_labels, on=['Id'], how='left')
```

```python
merged_df=merged_df.drop("Catcode",axis=1)
```

```python
y=merged_df["label"]
merged_df=merged_df.drop("label",axis=1)
merged_df
x=merged_df
```

```python
x["Beneficiary_Name"].isna().sum()
x["Beneficiary_Name"]=x["Beneficiary_Name"].fillna("Unknown")
# test_merged_df["Beneficiary_Name"]=test_merged_df["Beneficiary_Name"].fillna("Unknown")
```

```python
beneficiary_name=x.Beneficiary_Name.unique()
# test_beneficiary_name=test_merged_df.Beneficiary_Name.unique()
# beneficiary_name=(beneficiary_name+test_beneficiary_name).unique()
beneficiary_name=beneficiary_name.tolist()
beneficiary_name
```

```python
description=x.Description.unique()
# test_description=test_merged_df.Description.unique()
# description=(description+test_description).unique()
description=description.tolist()
description
```

```python
direction=x.Direction.unique()
direction=direction.tolist()
direction
```

# Polje nepotrebno jer ima jednu vrednost


```python
currency=x.Currency.unique()
currency=currency.tolist()
currency
#test_merged_df=test_merged_df.drop("Currency",axis=1)
x=x.drop("Currency",axis=1)
```

```python
kind=x.Kind.unique()
# test_kind=test_merged_df.Kind.unique()
# description=(description+test_kind).unique()
kind=kind.tolist()
kind
```

```python
from pandas.core.dtypes.dtypes import CategoricalDtype

beneficiary_name_cat=CategoricalDtype(categories=beneficiary_name,ordered=False)
description_cat=CategoricalDtype(categories=description,ordered=False)
kind_cat=CategoricalDtype(categories=kind,ordered=False)
direction_cat=CategoricalDtype(categories=direction,ordered=False)
x["Beneficiary_Name"]=x["Beneficiary_Name"].astype(beneficiary_name_cat)
x["Description"]=x["Description"].astype(description_cat)
x["Kind"]=x["Kind"].astype(kind_cat)
x["Direction"]=x["Direction"].astype(direction_cat)
```

```python
x['Date'] = pd.to_datetime(x['Date'])
x["Date"]=x['Date'].apply(lambda x:x.toordinal())
x["Mcc"]=np.nan_to_num(x["Mcc"],0)
x
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder , LabelEncoder , OrdinalEncoder

X_Encoded=pd.get_dummies(x, columns=["Beneficiary_Name","Description","Kind","Direction"])
X_Encoded
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_Encoded,y,test_size=0.2,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
X_norm=pd.DataFrame(X_train,columns=X_Encoded.columns)
X_norm.head()
AllData=scaler.fit_transform(X_Encoded)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score



lr=LogisticRegression()
lr.fit(X_train,y_train)


adaBoost=AdaBoostClassifier(n_estimators=100)
adaBoost.fit(X_train,y_train)
randomForest=RandomForestClassifier(n_estimators=100)
randomForest.fit(X_train,y_train)
```

```python
y_pred_lr=lr.predict(X_test)
y_pred_lr_prob=lr.predict_proba(X_test)
# y_pred_lr
y_pred_rf=randomForest.predict(X_test)
y_pred_ab=adaBoost.predict(X_test)
```

```python
rf_test=accuracy_score(y_test,y_pred_rf)
rf_test
```

```python
acc_ab_test=accuracy_score(y_test,y_pred_ab)
acc_ab_test
```

```python
lr_test=accuracy_score(y_test,y_pred_lr)
lr_test
```
